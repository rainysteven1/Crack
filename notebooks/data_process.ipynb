```python
import os

import pandas as pd

LOG_PATH = "../logs/train/runs"
CSV_PATH = "csv/version_0/metrics.csv"
csv_path = os.path.join(LOG_PATH, "2024-05-19_04-05-07", CSV_PATH)

csv_path
```

```python
ori_df = pd.read_csv(csv_path)
ori_df
```

```python
df_list = list(ori_df.groupby(["epoch", "step"]))
for _, df in df_list:
    df.ffill(inplace=True)
    df.bfill(inplace=True)
new_df = pd.concat([df for _, df in df_list])
new_df.drop_duplicates(inplace=True, keep="first")
new_df.reset_index(inplace=True, drop=True)
new_df
```

```python
val_acc_best = new_df["val/acc_best"].max()
new_df[new_df["val/acc"] == val_acc_best]
```
